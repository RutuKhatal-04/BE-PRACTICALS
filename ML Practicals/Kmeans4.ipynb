```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
data = iris.data  # Extract the feature data
```

```python
# StandardScaler is used to scale features to have a mean of 0 and standard deviation of 1
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)  # Scale the data
```

```python
# List to store Within-Cluster Sum of Squares (WCSS) values for different k values
# WCSS measures the sum of squared distances between each point and its cluster center
wcss = []
```

```python
# Using the elbow method to find the optimal number of clusters
# Loop through different cluster numbers (1 to 10)
for i in range(1, 11):
    # Initialize KMeans with current cluster count (n_clusters=i)
    # init='k-means++' selects initial cluster centers smartly to improve performance
    # max_iter=30 limits the number of iterations
    # n_init=10 specifies how many times the algorithm will run with different initializations
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=30, n_init=10)
    kmeans.fit(data_scaled)  # Fit KMeans to the scaled data
    wcss.append(kmeans.inertia_)  # Inertia is the WCSS value for the current k
```

```python
# Plot the elbow graph to determine the optimal number of clusters
plt.figure(figsize=(8, 6))
# Plot WCSS values against the number of clusters (k)
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal k')  # Set title for the plot
plt.xlabel('Number of Clusters')  # X-axis label
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')  # Y-axis label
plt.grid()  # Display grid for better readability
plt.show()  # Show the plot
```
